## 📊 Explanation of the Analytical Approach

### 🗂️ Data Preparation

I started with two datasets:

1. **💰 Marketing spend data** – daily spend for Google Search, Facebook, and Display, plus a promotion flag.  
2. **👥 Lead‑level data** – records of leads, qualified leads, and purchases with dates.

I merged them into a **daily master table** that aligned spend with the same day’s leads, qualified leads, and purchases.  
Days with no conversions were filled with **zero**.

To reduce daily noise and capture monthly patterns, I aggregated the data by **year and month**, creating a final analysis table of **31 monthly observations** (February 2022 – August 2024).  
This aggregation smoothed volatility and made linear relationships more discernible.

---

### 🔧 Building the Models

I built **three separate linear regression models**, each predicting a key business outcome:

- 📈 **Leads**  
- ✅ **Qualified leads**  
- 💳 **Purchases** (primary outcome)

The **independent variables** were:
- Google Search spend  
- Facebook spend  
- Display spend  
- Promotion flag (0/1)

I split the monthly data into **training (75%)** and **test (25%)** sets, fitted each model on the training set, and evaluated them using **R², RMSE, and MAE**.

---

### 🧠 Why Linear Regression?

I chose linear regression for several practical and interpretability reasons:

| Reason | Explanation |
|--------|-------------|
| 📐 **Interpretability of marginal impact** | Coefficients directly show the estimated change in the outcome per unit increase in channel spend, holding other factors constant. This answers: *“Which channel has the greatest marginal business impact?”*  
  *Example:* Display’s coefficient for purchases (`0.001463`) means each extra dollar spent on Display is associated with a **0.001463 increase** in purchases. |
| 🧾 **Simplicity & transparency** | Easy to explain to non‑technical stakeholders. Coefficients can be presented in a simple table, making the recommendation clear and defensible. |
| 🎯 **Adequate for exploratory marginal analysis** | Even though the purchases model had a negative R² (poor predictive power), the coefficients still provide a **relative ranking** of marginal effects. The goal was not high‑accuracy prediction, but comparing channels. |
| 📊 **Baseline for future models** | Linear regression serves as a natural starting point. It can later be compared to non‑linear or time‑series models (e.g., SARIMAX, Bayesian structural time series) if causal inference is needed. |
| 📈 **Aligns with marketing mix modelling** | Many industry budget allocation approaches use linear (or log‑linear) marginal response functions, especially with monthly aggregated data. |

---

### 📋 How I Used the Output

After obtaining the coefficients, I compared them across channels, prioritising:

1. **💳 Purchases** (most important business outcome)  
2. **✅ Qualified leads**  

**Key findings:**
- **Display** had the highest positive coefficient for **purchases (0.001463)** and also for **qualified leads (0.006456)**.  
- **Facebook** had smaller positive effects on both.  
- **Google Search** had negative coefficients for qualified leads and purchases.

👉 **Recommendation:** **Display Advertising** should receive the incremental budget.

---

### ⚠️ Limitations Acknowledged

- The analysis treats **associations as marginal effects**, not causal relationships, because the data lacks:
  - Channel‑level lead attribution  
  - Revenue or customer lifetime value (CLV)   
- Linear regression assumes **linearity and independence of errors** – violations may exist.  
- However, for the purpose of **ranking marginal impact** to support decision‑making, the approach remains useful and actionable.

## 📈 How to Improve This Analysis

To make the recommendation **stronger, more actionable, and causally robust**, the following additional data, metrics, and analytical approaches would be valuable.

---

### 🔍 Additional Data Needed

| Data Type | Why It Matters (Explained Simply) |
|-----------|-----------------------------------|
| **💰 Income (Revenue) per purchase** | Currently, we treat every purchase equally, but in reality, purchases have different values. A small sale and a large enterprise contract are not the same. Knowing the **revenue** generated by each purchase would allow us to calculate **ROMI (Return on Marketing Investment)** – that is, *how many dollars we earn back for every dollar spent*. Without revenue, we might over‑value low‑value purchases. |
| **💸 Cost per channel (beyond just spend)** | Our current “spend” is what we pay the platform (Google, Facebook, etc.). But there are often additional costs: agency management fees, creative production, or software costs. Including these would give a truer picture of **marginal profitability** per channel. |
| **👥 User / customer identifiers** | If we could track individual customers over time (e.g., using a customer ID), we could see whether a lead made **repeat purchases** and calculate **Customer Lifetime Value (CLV)**. A channel that attracts one‑time buyers might be less valuable than a channel that brings loyal, long‑term customers. |
| **🎯 Channel‑level lead attribution** | Currently, we only know that *some* leads were generated on a given day, but we don’t know *which channel* brought each lead. Without that, we cannot directly link a specific dollar spent to a specific lead or purchase. With attribution, we could build much more precise models (e.g., logistic regression) that predict conversion probability by channel. This would also allow us to control for other factors (like seasonality) more effectively. |
| **📅 Daily data with more context** | We aggregated data to monthly level to reduce noise, but daily data (with extra variables like day of week, holidays, or competitor activity) would allow us to capture short‑term effects, such as a promotion that lasts only a few days. It would also help us see if weekends behave differently from weekdays. |

---

### 📊 Key Metric: ROMI (Return on Marketing Investment)

**What is ROMI?**  
ROMI measures the **financial return** from marketing spending. It answers the question: *“For every $1 we spend on a channel, how many $ do we get back in revenue?”*

**Formula:**  
`ROMI = (Incremental Revenue Attributed to Marketing – Marketing Cost) / Marketing Cost`

**Why ROMI is critical:**  
- It focuses on **profit**, not just on the number of purchases or leads.  
- It allows **fair comparison** between channels with very different costs and revenue profiles.  
- It accounts for diminishing returns (if we already spend a lot, the next dollar might bring less revenue).

> ⚠️ In this analysis, **ROMI could not be calculated** because we lacked:
> 1. Revenue per purchase  
> 2. Channel‑level attribution  

Without ROMI, we can only compare purchase *counts*, which is less informative. A channel that brings 10 small purchases ($100 each) might be worse than a channel that brings 5 large purchases ($1,000 each), even if the cost is the same. ROMI would reveal that.

---

### 📋 Summary of Recommended Enhancements

- **Add revenue data** to calculate ROMI – the most important metric for budget decisions.  
- **Add user IDs** to track lifetime value and repeat purchases.  
- **Add channel attribution** so we can directly link spend to outcomes.   
- **Include all costs** (agency fees, production) to get true marginal profitability.

With these additions, we could move from **estimating associations** (e.g., “Display seems to generate more purchases”) to **optimising for profit** (e.g., “Display delivers $5 of revenue per $1 spent, which is the best ROMI”). That would make the recommendation dramatically more reliable and actionable.
